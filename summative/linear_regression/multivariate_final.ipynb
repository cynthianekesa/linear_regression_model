1. Correlation Analysis:

     - Fixed acidity and citric acid (0.7)
     - Fixed acidity and density (0.7)
     - Free SO2 and total SO2 (0.7)
     - Quality and goodquality (0.8)

     - Fixed acidity and pH (-0.7)

2. Features to Consider Dropping:
   a. Total SO2: Highly correlated with free SO2 (0.7). free SO2 is more directly relevant.
   b. Density: Moderately correlated with fixed acidity (0.7) and shows low correlation with quality. Fixed acidity is more informative.
   c. pH: Strongly correlated with fixed acidity (-0.7). Fixed acidity is more preferred.
   d. Citric acid: Strongly correlated with fixed acidity (0.7). fixed acidity is more preferred as it shows slightly higher correlations with other features.

3. Features to Keep:
   - Fixed acidity
   - Volatile acidity
   - Residual sugar
   - Chlorides
   - Free SO2/total SO2
   - Sulphates
   - Alcohol

Quality and goodquality are highly correlated (0.8), suggesting they are measuring similar aspects. Therefore I am choosing quality as the target variable and goodquality as the attribute that will help in interpreting prediction results.
Alcohol shows the most consistent positive correlation with quality measures, making it an important feature to retain.

```python
# temporarily removing the goodquality column
X = a.drop('goodquality', axis=1)
y = a['goodquality']
```

```python
# temporarily dropping the quality column which is the target variable and doesnt need to be permanently dropped
X = a.drop('quality', axis=1)
y = a['quality']
```

```python
# the data has now been split into features and targets
a.columns
```

```python
# scaling the data to be in the same range
from sklearn.preprocessing import MinMaxScaler
# create an object of the scaler
scaler = MinMaxScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
# spliting the data into Training and test
from sklearn.model_selection import train_test_split


# split the data

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# building the model
from sklearn.linear_model import LinearRegression

# make a model instance
model = LinearRegression()

# fit the model
model.fit(X_train, y_train)
```

```python
# making predictions and determinimg model accurancy
from sklearn.metrics import mean_absolute_error

# make some Prediction in the model
y_pred = model.predict(X_test)

# plot the predicted values with the test values
mean_abo_err = mean_absolute_error(y_test, y_pred)
print(mean_abo_err)

# mean error score
mean_error_score = np.sqrt(mean_absolute_error(y_test, y_pred))
print(mean_error_score)


# print the model score to predict the values
print(model.score(X_test, y_test) * 100)
```

```python
# save the model using pickle
import pickle as pk

# Saving model to pickle file
with open("./model.pkl", "wb") as file: # file is a variable for storing the newly created file
    pk.dump(model, file)
```

```python
import asyncio
import uvicorn
from typing import Annotated
from fastapi import FastAPI, Depends, HTTPException, status, Path
from pydantic import BaseModel, Field

app = FastAPI()
from pydantic import BaseModel, Field

class WineQRequest(BaseModel):
    fixed_acidity: float = Field(gt=0, lt=10000)
    volatile_acidity: float = Field(gt=0, lt=10000)
    residual_sugar: float = Field(gt=0, lt=10000)
    chlorides: float = Field(gt=0, lt=10000)
    total_SO2: float = Field(gt=0, lt=10000)
    sulphates: float = Field(gt=0, lt=10000)
    alcohol: float = Field(gt=0, lt=10000)



@app.get("/", status_code=status.HTTP_200_OK)
async def get_hello():
    return {"hello": "summative yeey"}

@app.post('/predict', status_code=status.HTTP_200_OK)
async def make_prediction(wineq_request: WineQRequest):
    try:
        single_row = [[wineq_request.fixed_acidity, wineq_request.volatile_acidity, wineq_request.residual_sugar, wineq_request.chlorides, wineq_request.total_SO2, wineq_request.sulphates, wineq_request.alcohol]]
        new_data_scaled = scaler.transform(single_row)
        new_value = model.predict(new_data_scaled)
        return {"predicted Quality ": new_value[0]}
    except:
        raise HTTPException(status_code=500, detail="Something went wrong.")



if __name__ == "__main__":
    config = uvicorn.Config(app)
    server = uvicorn.Server(config)
    await server.serve()
```
